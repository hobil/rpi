```python
import multiprocessing, time, signal
```

```python
def x(): 
    while True:
        time.sleep(1)
        print(1)  
```

```python
p = multiprocessing.Process(target=x)
```

```python
print(p, p.is_alive())
```

```python
p.start()
```

```python
print(p, p.is_alive())
```

```python
p.join(0.5)
```

```python
p.terminate()
```

```python
print(p, p.is_alive())
```

```python
p.exitcode == -signal.SIGTERM
```

```python
p.start()
```
